```python
def num_giv(*args):
    # Filter only numeric values
    value = [x for x in args if isinstance(x, (int, float))]
    
    # If no numeric value, return error
    if not value:
        return "error"
    
    # Return average of numeric values
    return sum(value) / len(value)

# Function call
print(num_giv(2))               # Output: 2.0
print(num_giv(2, 'a', 5, 3.5))  # Output: 3.5
print(num_giv('x', None))       # Output: error
```

```python
class Account:
    def __init__(self,account_num,balance):
        self.account_num=account_num
        self.balance=balance
    def deposit(self):
        return "deposit kar wa suniya"
    def withdraw(self):
        return "pasianhi na"
    def show_balance(self):
        return self.balance
        
class SavingAccount(Account):
    def __init__(self,interest_rate,account_num,balance):
         super().__init__(account_num, balance) 
         self.interest_rate=interest_rate
        
    def add_interest(self):
        return self.interest_rate
        
        
class CurrentAccount(Account):
    def __init__(self,overdraft_limit,account_num,balance):
         super().__init__(account_num, balance) 
         self.overdraft_limit=overdraft_limit
        
a=Account(121956,1450000)
sa=SavingAccount("ali",121567,345667)
ca=CurrentAccount(121956,345667,50000000)
a.deposit()
a.show_balance()
sa.withdraw()
ca.withdraw()
```

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def show_info(self):
        print(f" the name is {self.name} and price {self.price}")
        
class Book(Product):
    def __init__(self,name,price,author):
        super().__init__(name,price)
        self.author=author
    def show_info(self):
        super().show_info()
        print(f"the author os {self.author}")
class Electronics(Product):
    def __init__(self,name,price,warranty):
        super().__init__(name,price)
        self.warranty=warranty
    def show_info(self):
        super().show_info()
        print(f"the warranty {self.warranty}")
class Mobile(Electronics):
    def __init__(self,price,name,warranty,battery_life):
        super().__init__(name,price,warranty)
        self.battery_life=battery_life
    def show_info(self):
        super().show_info()
        print(f" the battery is {self.battery_life}")
        
p = Product("Dryer", 13000)
b = Book("Harry Potter", 1200, "J.K. Rowling")
e = Electronics("Dispenser", 12000, "1 year")
m = Mobile("iPhone 14", 250000, "1 year", 20)
p.show_info()
print("---")
b.show_info()
print("---")
e.show_info()
print("---")
m.show_info()

        
```
