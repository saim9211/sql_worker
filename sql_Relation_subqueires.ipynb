```python
import sqlite3

conn = sqlite3.connect("library.db")
cur = conn.cursor()

# Enable foreign keys
cur.execute("PRAGMA foreign_keys = ON;")

# Drop tables if already exist
cur.executescript("""
DROP TABLE IF EXISTS purchase;
DROP TABLE IF EXISTS books;
DROP TABLE IF EXISTS profile;
""")

# Create profile table
cur.execute("""
CREATE TABLE profile (
    user_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT UNIQUE,
    role TEXT
);
""")

# Create books table
cur.execute("""
CREATE TABLE books (
    book_id INTEGER PRIMARY KEY,
    title TEXT NOT NULL,
    author TEXT,
    price INTEGER,
    stock INTEGER
);
""")

# Create purchase table
cur.execute("""
CREATE TABLE purchase (
    purchase_id INTEGER PRIMARY KEY,
    user_id INTEGER,
    book_id INTEGER,
    purchase_date TEXT,
    FOREIGN KEY (user_id) REFERENCES profile(user_id),
    FOREIGN KEY (book_id) REFERENCES books(book_id)
);
""")

# Insert data into profile
cur.executescript("""
INSERT INTO profile VALUES (1,'Ali','ali@gmail.com','student');
INSERT INTO profile VALUES (2,'Sara','sara@gmail.com','student');
INSERT INTO profile VALUES (3,'Admin','admin@gmail.com','admin');
""")

# Insert data into books
cur.executescript("""
INSERT INTO books VALUES (101,'Python Basics','John Smith',1500,5);
INSERT INTO books VALUES (102,'Data Science','Andrew Lee',2000,2);
INSERT INTO books VALUES (103,'Machine Learning','Tom Hardy',2500,0);
""")

# Insert data into purchase
cur.executescript("""
INSERT INTO purchase VALUES (1,1,101,'2026-01-01');
INSERT INTO purchase VALUES (2,2,102,'2026-01-02');
""")

conn.commit()


print("✅ Library dataset created successfully")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Table 1: Customers
cursor.execute("""
CREATE TABLE customers (
    id INTEGER PRIMARY KEY,
    name TEXT,
    region TEXT
)
""")

cursor.executemany("""
INSERT INTO customers (id, name, region) VALUES (?, ?, ?)
""", [
    (1, 'Alice', 'East'),
    (2, 'Bob', 'East'),
    (3, 'Carol', 'West'),
    (4, 'Dave', 'West'),
    (5, 'Eve', 'North'),
    (6, 'Frank', 'North')
])

# Table 2: Products
cursor.execute("""
CREATE TABLE products (
    id INTEGER PRIMARY KEY,
    name TEXT,
    price INTEGER
)
""")

cursor.executemany("""
INSERT INTO products (id, name, price) VALUES (?, ?, ?)
""", [
    (1, 'Laptop', 1000),
    (2, 'Phone', 500),
    (3, 'Tablet', 700)
])

# Table 3: Orders
cursor.execute("""
CREATE TABLE orders (
    id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    FOREIGN KEY(customer_id) REFERENCES customers(id),
    FOREIGN KEY(product_id) REFERENCES products(id)
)
""")

cursor.executemany("""
INSERT INTO orders (customer_id, product_id, quantity) VALUES (?, ?, ?)
""", [
    (1, 1, 1),   # Alice buys 1 Laptop
    (1, 2, 2),   # Alice buys 2 Phones
    (2, 2, 1),   # Bob buys 1 Phone
    (3, 3, 2),   # Carol buys 2 Tablets
    (4, 2, 3),   # Dave buys 3 Phones
    (5, 1, 1),   # Eve buys 1 Laptop
    (6, 3, 1)    # Frank buys 1 Tablet
])

conn.commit()
```

```python
query = """
SELECT 
    c.name AS customer,
    c.region,
    SUM(p.price * o.quantity) OVER(PARTITION BY c.name) AS total_spent
FROM orders o
JOIN customers c ON o.customer_id = c.id
JOIN products p ON o.product_id = p.id;
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
SELECT 
    c.name AS customer,
    c.region,
    SUM(p.price * o.quantity) AS amount,
    RANK() OVER(PARTITION BY c.region ORDER BY SUM(p.price * o.quantity) DESC) AS rank_in_region
FROM orders o
JOIN customers c ON o.customer_id = c.id
JOIN products p ON o.product_id = p.id
GROUP BY c.name, c.region;
"""

df = pd.read_sql_query(query, conn)
df
```
