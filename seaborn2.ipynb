```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
tips = sns.load_dataset("tips")

sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    row="sex",          # separate rows for male/female
    col="day",          # separate columns for days
    aspect=0.5,         # width/height ratio of plots
    scatter_kws={"s": 20, "edgecolor": "black", "color": "blue"},  # scatter settings
    line_kws={"linewidth": 1.4, "color": "red"}                    # regression line settings
)

plt.show()
```
